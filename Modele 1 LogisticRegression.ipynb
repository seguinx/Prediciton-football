Data Import

```python
import pandas as pd
df = pd.read_csv('football.csv')
df_comp = pd.read_csv("competition.csv")
```

Data Cleaning

```python
df.info()
```

```python
df
```

Drop Unnecessary Columns

```python

df = df.drop(columns=['date','id','home_team','away_team','match_type','rank_change_away'])
```

Impute Missing Values

```python
df.isna().sum()
```

```python
df['home_goals_suf_mean'] = df['home_goals_suf_mean'].fillna(df['home_goals_suf_mean'].median())
df['home_points_mean'] = df['home_points_mean'].fillna(df['home_points_mean'].median())
df['away_goals_suf_mean'] = df['away_goals_suf_mean'].fillna(df['away_goals_suf_mean'].median())
```

```python
df
```

Split Data

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns=['target'])
y=df['target']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)


```

Scale Data for Fair Comparison

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

Train Ridge & Lasso Models

```python
from sklearn.linear_model import Ridge, Lasso

ridge = Ridge(alpha=0.1)
lasso = Lasso(alpha=0.1, max_iter=1000)

ridge.fit(x_train, y_train)
lasso.fit(x_train, y_train)
```

Compare Ridge vs Lasso Performance

```python
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
y_pred_ridge = ridge.predict(x_test)
y_pred_lasso = lasso.predict(x_test)

y_pred_ridge_bin = (y_pred_ridge > 0.5).astype(int)
y_pred_lasso_bin = (y_pred_lasso > 0.5).astype(int)


rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge_bin))
r2_ridge = r2_score(y_test, y_pred_ridge_bin)
rmse_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso_bin))
r2_lasso = r2_score(y_test, y_pred_lasso_bin)

print("RIDGE: RMSE =", round(rmse_ridge, 3), "R² =", round(r2_ridge, 3))
print("LASSO: RMSE =", round(rmse_lasso, 3), "R² =", round(r2_lasso, 3))
print("Ridge Accuracy TEST:", (y_pred_ridge_bin == y_test).mean())
print("LASSO Accuracy TEST:", (y_pred_lasso_bin == y_test).mean())
```

Improve Ridge Model Performance

```python
top10 = pd.DataFrame({
    'Colonne': df.columns[1:], 
    'Coef': ridge.coef_
}).sort_values('Coef', key=abs, ascending=False).head(10)

print(top10['Colonne'].tolist())
```

```python
from sklearn.metrics import accuracy_score
# 1️⃣ Sélectionne TES 10 colonnes
top10_cols = ['away_rank_mean_l5', 'home_rank_mean_l5', 'away_goals_suf_mean_l5', 
              'home_goals_mean_l5', 'home_goals_suf_mean_l5', 'away_goals_mean_l5', 
              'home_rank_mean', 'home_points_mean_l5', 'away_points_mean_l5', 'away_rank_mean']

x_top10 = df[top10_cols]

# 2️⃣ Split
x_train_top10, x_test_top10, y_train, y_test = train_test_split(x_top10, y, test_size=0.2)

x_train_top10, x_test_top10, y_train, y_test = train_test_split(x_top10, y, test_size=0.2)

scaler_top10 = StandardScaler()
x_train_top10_scaled = scaler_top10.fit_transform(x_train_top10)
x_test_top10_scaled = scaler_top10.transform(x_test_top10)

# Ridge TOP 10
ridge_top10 = Ridge(alpha=0.001,random_state=1)
ridge_top10.fit(x_train_top10_scaled, y_train)

# TEST
y_pred_top10 = ridge_top10.predict(x_test_top10_scaled)
print("Accuracy TOP 10:", accuracy_score(y_test, (y_pred_top10 > 0.5).astype(int)))

```
